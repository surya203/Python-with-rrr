```python
class movie:
    def movie_collection(self,collection):
        self.collection = collection
        print("The total collection amount is",collection,"cr")
        
    def movie_promotions(self, promotions):
        self.promotions = promotions
        print("The total promotions amoumt is",promotions,"cr")
        
    def movie_making_cost(self, making):
        self.making = making
        print("The total making amount is",making,"cr")
```

```python
class movies:
    def total_profit(self,making_cost,promotions_cost,collections):
        self.making_cost = making_cost
        self.promotions_cost = promotions_cost
        self.collections = collections
        profit = collections - making_cost - promotions_cost
        return profit
        
```

```python
rrr = movies()
```

```python
rrr.total_profit(500,10,1200)
```

```python
kgf2 = movies()
```

```python
kgf2.total_profit(200,5,1200)
```
